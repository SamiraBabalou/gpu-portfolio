# Import libraries
import numpy as np
import cupy as cp  # GPU array library

# Create arrays
N = 1000000
a = cp.arange(N, dtype=cp.float32)
b = cp.arange(N, dtype=cp.float32)

# GPU addition
c = a + b

# Move to CPU and verify
c_cpu = cp.asnumpy(c)
print("First 5 elements:", c_cpu[:5])
